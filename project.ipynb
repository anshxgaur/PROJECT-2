Q1:Use yfinance to Extract Tesla Stock Data

```python
# Import the yfinance library
import yfinance as yf

# Define the ticker symbol for Tesla
tickerSymbol = 'TSLA'

# Create a ticker object using yfinance
tickerData = yf.Ticker(tickerSymbol)

# Extract historical stock data for the maximum available period
tickerDf = tickerData.history(period='max')

# Display the first five rows of the data
tickerDf.head()
```

Q2:Webscraping Tesla Revenue Data from Macrotrends

```python
# Import required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Send a GET request to the webpage
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table')

# Extract the Tesla revenue table
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        rows = table.find_all('tr')
        revenue_data = []
        for row in rows[1:]:
            cols = row.find_all('td')
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                revenue_data.append([date, revenue])
        break

# Create a DataFrame from the scraped data
tesla_revenue = pd.DataFrame(revenue_data, columns=["Date", "Revenue"])

# Display the last five rows
tesla_revenue.tail()
```

Q5: Plot Tesla Stock Graph

```python
import matplotlib.pyplot as plt

def make_graph(data, stock, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label=stock, color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Assuming tesla_df is already defined and has 'Date' and 'Close' columns
make_graph(tesla_df, 'TSLA', 'Tesla Stock Price Over Time')
```
